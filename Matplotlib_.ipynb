```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
plt.scatter(tips['total_bill'],tips['tip'],color='red',s=tips['size']*10) # s is used to show other piece of information, how is tip related to family size
```

```python
plt.plot(tips['total_bill'],tips['tip'],'o') # another way of plotting scatterplot, not so functional
```

```python
plt.bar(tips['sex'],tips['tip'])
```

```python
plt.barh(tips['sex'],tips['tip']) # when number of categories are more
```

```python
avg = tips.groupby('sex')['tip'].mean() # avg is now a pandas series
plt.bar(avg.index, avg.values)
plt.xlabel("Sex")
plt.ylabel("Average Tip")
plt.title("Average Tip by Sex")
```

```python
plt.hist(tips['tip'])
```

```python
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Sales': [150, 200, 300, 100, 250],
    'Profit': [20, 35, 50, 10, 40],
    'Quantity': [30, 45, 60, 20, 50]
}

df3 = pd.DataFrame(data)

print(df3)
```

```python
petal=sns.load_dataset("iris")
```

```python
petal.head()
```

```python
plt.scatter(petal['petal_length'],petal['petal_width'],c=petal['species'].astype('category').cat.codes,cmap='viridis')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs Petal Width by Species')
plt.colorbar(label='Species Code') # Add a color bar to show the mapping of codes to colors
```

```python
# First subplot
plt.subplot(1, 3, 1)   # (1 row, 2 columns, 1st plot)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("Plot 1")

# Second subplot
plt.subplot(1, 3, 2)   # (1 row, 2 columns, 2nd plot)
plt.plot([1, 2, 3], [1, 2, 3])
plt.title("Plot 2")

plt.subplot(1, 3, 3)   # (1 row, 2 columns, 2nd plot)
plt.plot([1, 2, 3], [1, 2, 3])
plt.title("Plot 3")

plt.show()
```

```python
fig, axes = plt.subplots(2, 2)  # 2 rows  2 cols = 4 plots

axes[0, 0].plot([1,2,3], [1,4,9])
axes[0, 0].set_title("Top-Left")

axes[0, 1].plot([1,2,3], [1,2,3])
axes[0, 1].set_title("Top-Right")

axes[1, 0].bar([1,2,3], [3,2,1])
axes[1, 0].set_title("Bottom-Left")

axes[1, 1].hist([1,1,2,3,3,3,4])
axes[1, 1].set_title("Bottom-Right")

plt.tight_layout()  # adjust spacing
plt.show()
```
