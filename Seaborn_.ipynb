```python
import numpy as np
import  pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
# scatter plot-> axes level function
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size') # add extra layer of information, automatic legend creation
```

```python
import plotly.express as px
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp')  # axis level function
```

```python
temp_df2=gap[gap['continent'].isin(['Asia','Europe'])]
```

```python
sns.relplot(data=temp_df2,x='year',y='lifeExp',hue='continent',kind='line') # figure level function
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',col='time',row='day') # facet plot-> multiple plots plotted based on diff categorical columns # only for figure level functions  # col_wrap
```

```python
sns.histplot(data=tips,x='total_bill') #axes level function
sns.displot(data=tips, x='total_bill', kind='hist') # figure level function
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',fill=True) # figure level function
sns.rugplot(data=tips,x='total_bill') # axeslevel function
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',hue='sex',col='day',element='step') # faceting with distribution plots
```

```python
#count plot through logic of hist -> works on categorical columns, returns count of each category
sns.displot(data=tips,x='day',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

```python
temp_df3 = gap.pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(data=temp_df3) # axes level function
```

```python
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='summer')
```

```python
iris = px.data.iris()
iris
```

```python
# cluster map is similar to heatmap but internally performs clustering on data, dendograms
sns.clustermap(data=iris.iloc[:,[0,1,2,3]],cmap='summer')
# A dendrogram is a diagram that shows the hierarchical merging of data points into clusters, with branch height representing the distance (similarity) between them. Its widely used in hierarchical clustering to understand data structure.
```

```python
sns.boxplot(data=tips,x='total_bill') #univariate analysis
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
sns.countplot(data=tips,x='sex',hue='day')
```

```python
sns.regplot(data=tips,x='total_bill',y='tip')
# axes level function
#provides scatterplot + fits a regression line
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='day',hue='sex')
# figure level function
#Adds support for faceting with row and col parameters
#Good when you want to compare regression lines across categories
```

```python
sns.residplot(data=tips,x='total_bill',y='tip')
#axes level function
#plots the difference between actual y values and predicted y values (residuals)
#useful in checking regression assumptions
```

```python
sns.pairplot(data=iris,hue='species')
```

```python
# pair grid
g = sns.PairGrid(data=iris,hue='species')
# g.map
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data=iris.drop(columns=['species_id']),hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris.drop(columns=['species_id']),hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist',hue='sex')
```

```python
g = sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.violinplot)
```
